```python
# Text cleaning function
def clean_text(text):
    
    text = re.sub(r'RT @[A-Za-z0-9_]+', '', text)
    # Remove words starting with '@'
    text = re.sub(r'@\w+', '', text)
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    # Remove non-alphabetic characters except spaces
    text = re.sub(r'[^A-Za-z\s]', '', text)
    # Convert to lower case
    text = text.lower().strip()
    
    # Remove meaningless words (example: long sequences of characters)
    words = text.split()
    cleaned_words = [word for word in words if not re.match(r'^[a-z]{10,}$', word)]
    
    return ' '.join(cleaned_words)
```

```python
df['text'] = df.apply(lambda x: clean_text(x['text']), axis=1)
df
```

```python
tokenize = RegexpTokenizer(r'\w+')
df['text'] = df.apply(lambda x: tokenize.tokenize(x['text']), axis=1)
df
```

```python
# Replace bahasa gaul, Handle elongation = Normal Text
from indoNLP.preprocessing import pipeline, replace_word_elongation, replace_slang
pipe = pipeline([replace_word_elongation, replace_slang])
def non_alay(sentence):
    new_sentence = []
    for i in sentence:
        new_sentence.append(pipe(i))
    return(new_sentence)
```

```python
df['text'] = df.apply(lambda x: non_alay(x['text']), axis=1)
df
```

```python
import stanza

nlp = stanza.Pipeline(lang='id', processors='tokenize,mwt,pos,lemma')
```

```python
doc = nlp('mengandung perkemahan')
print(*[f'word: {word.text+" "}\tlemma: {word.lemma}' for sent in doc.sentences for word in sent.words], sep='\n')
```

```python
# Lemmanization
def lemmanization(sentence):
    new_sentence = []
    for i in sentence:
        new_sentence.append(nlp(i).sentences[0].words[0].lemma)
    
    return(new_sentence)
```

```python
res=df['text'].apply(lemmanization)
res
```
